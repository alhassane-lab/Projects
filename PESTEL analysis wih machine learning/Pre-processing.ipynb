## Indicateurs liés à la disponibilité alimentaire

```python
# disponibilité alimentaire uniquement sur les volailles, source : FAO
dispo = pd.read_csv("./data/dispo_en.csv")
dispo.head()
```

```python
dispo = dispo[['Area', 'Element', 'Item', 'Year', 'Value']]
dispo2 = dispo.pivot_table(index=["Area", "Year"], columns=["Element"])
dispo2.head()
```

```python
t =dispo.pivot_table(index=["Area", "Year"], columns=["Element"])
r=[]
for i in t.columns:
    r.append(i[1])
r
```

```python
dispo2.columns=['Domestic supply quantity',
 'Export Quantity',
 'Fat supply quantity (g/capita/day)',
 'Food',
 'Food supply (kcal/capita/day)',
 'Food supply quantity (kg/capita/yr)',
 'Import Quantity',
 'Losses',
 'Other uses (non-food)',
 'Processing',
 'Production',
 'Protein supply quantity (g/capita/day)',
 'Residuals',
 'Stock Variation',
 'Tourist consumption']
```

```python
dispo2 = dispo2.reset_index()
dispo2
```

```python
# On garde les données de l'année 2019
dispo2019 = dispo2[dispo2["Year"] == 2019]
dispo2019 = dispo2019.fillna(0)
dispo2019.head()
```

```python
dispo2019["Area"].replace(["C?te d'Ivoire"],["Côte d'Ivoire"], inplace=True)
```

```python
dispo2019.info()
```

## Indicateurs macro économiques

```python
gdp = pd.read_csv("./data/FAO_GDP.csv")
gdp.head()
```

```python
gdp["Item"] += " (" + gdp["Element"] + ")"
gdp2019= gdp[gdp["Year"] ==2019]
gdp2019 = pd.pivot_table(gdp2019, index="Area", columns="Item",values="Value")
gdp2019.columns =['GDP growth per capita (US$)',
       'GDP growth (US$)',
       'GDP per capita (US$)',
       'GDP (US$)']
gdp2019 = gdp2019.reset_index()
```

```python
gdp2019["Area"].replace(["C?te d'Ivoire"],["Côte d'Ivoire"], inplace=True)
```

```python
gdp2019.head()
```

<div class="alert-info">
    <p style="text-align:center"><strong> Infos</strong> :</p>
    <p> Pour la suite de l'analyse, nous allons retenir les deux indiateurs macro-économiques suivants : </p>
    <ul>
      <li> <strong>GDP per capita  </strong> afin d'avoirun aperçu sur la production de la richesse par habitant pour chaque pays.</li>
      <li> <strong>GDP growth per capita  </strong> afin d'avoir un aperçu sur la croissance du pib par rapport à l'année d'avant. </li>
    </ul>
</div>
<br>

```python
gdp2019 = gdp2019.drop(columns=["GDP growth (US$)", "GDP (US$)"])
gdp2019.head()
```

```python
df = dispo2019.merge(gdp2019, on="Area", how="left")
df.head()
```

```python
df.loc[df["Area"]== "China, Taiwan Province of", ["GDP growth per capita (US$)","GDP per capita (US$)"] ] = df.loc[df["Area"]== "China, Macao SAR", ["GDP growth per capita (US$)","GDP per capita (US$)"]].values
df[df["Area"].str.contains("China")]
```

### 2.3. Indicateurs sociaux

```python
pop = pd.read_csv("./data/FAO_POP.csv")
pop.head()
```

```python
pop = pop[["Area", "Value"]]
pop.rename(columns={"2017":"Population (2017)","2018":"Population (2018)"}, inplace=True)
pop["Area"].replace(["C?te d'Ivoire"],["Côte d'Ivoire"], inplace=True)
```

```python
pop = pd.read_csv("FAOSTAT_data_6-2-2022 (1).csv")
pop = pop.pivot_table(index="Area",values="Value", columns="Year", aggfunc="sum")
pop.columns=["Population (2008)","Population (2009)","Population (2010)","Population (2011)","Population (2012)","Population (2013)","Population (2014)","Population (2015)","Population (2016)","Population (2017)","Population (2018)"]
pop = pop.reset_index()
pop["Area"].replace(["C?te d'Ivoire"],["Côte d'Ivoire"], inplace=True)
```

#### Le taux moyen de croissance de la population entre 2013 et 2018

```python
pop["Avg. population growth (annual %)"] = round(((pop["Population (2018)"]/pop["Population (2013)"])**(1/5)*100)-100,2)
pop.head()
```

```python
df = df.merge(pop[["Area","Avg. population growth (annual %)"]], on="Area", how="inner")
```

```python
political = pd.read_csv("./data/PoliticalStability.csv")
political = political[political["Year"]==2018]
political.rename(columns={"Country":"Area"}, inplace=True)
display(political.head())
political.info()
```

```python
p = df.merge(political[["Area","Political_Stability"]],on="Area", how="outer", indicator=True)
p[p["_merge"] == "left_only"]
```

```python
df = df.merge(political[["Area","Political_Stability"]],on="Area", how="inner")
display(df.head())
df.info()
```

    <ul>
      <li> <strong> Avg. population growth  </strong> c'est le taux de croissance moyen par an de la population sur cinq ans (2013 à 2018) </li>
      <li> <strong> Political stability </strong> afin d'avoir un aperçu sur la stabilité politique de chaque pays </li>
    </ul>
</div>

```python
region = pd.read_csv("./data/country_grp.csv")
region.rename(columns={"Country":"Area", "Country Group":"Region"}, inplace=True)
region = region[region["Region"].isin(['Africa', 'Americas', 'Asia', 'Europe','Oceania',])]
df = df.merge(region[["Area", "Region"]], on="Area", how="left")
df.loc[df["Area"]== "Turkey", ["Region"]] = "Europe"
```

```python
UE = ["Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czechia", "Denmark", "Estonia", "Finland", "France", "Germany", "Greece", "Hungary", "Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg", "Malta", "Netherlands", "Poland", "Portugal", "Romania", "Slovakia", "Slovenia", "Spain", "Sweden"]

df["Union european"] = 0
df["Union european"][df["Area"].isin(UE)] = 1
```
